Hay N cubos dispuestos en fila. Cada cubo está vacío o contiene
una pelota. Los depósitos se especifican como una cadena
que consta de caracteres "." (cubo vacío) y "B" (cubo con una
pelota). Por ejemplo, para depósitos = "B.BB.B..B", la fila de

En un movimiento puedes sacar la pelota de cualquier cubo y colocarla
en otro cubo (vacío). Tu objetivo es organizar las bolas para crear
una secuencia alterna de cubos llenos y vacíos. En otras palabras,
la distancia entre dos bolas consecutivas debe ser igual a 2.
Tenga en cuenta que la secuencia puede comenzar en cualquier cubo.
Por ejemplo, en la siguiente figura, las bolas están colocadas correctamente:   ¿Cuál es el número mínimo de movimientos necesarios para crear
una secuencia correcta de bolas en cubos?
Escribe una función:
solución def (cubos)
que, dada una cadena de cubos de longitud N, devuelve el número
mínimo de movimientos necesarios para crear la secuencia descrita.
Si es imposible crear una secuencia correcta, devuelve -1.
Ejemplos:
1. Dados los depósitos"..B... B.BB", la función debería devolver 2.

2. Dados los depósitos = "BB.B.BBB...", la función debería devolver 4.

3. Dados los depósitos = ".BBB.B", la función debería devolver
-1 porque es imposible crear una secuencia correcta.

4. Dados los cubos "......B.B", la función debería devolver
0. No es necesario mover ninguna bola porque la secuencia
ya es correcta.

```python
def solution(cubos):
    N = len(cubos)
    movimientos = 0

    # Verificar si ya es una secuencia correcta
    for i in range(N):
        if i % 2 == 0 and cubos[i] != 'B':
            continue
        elif i % 2 == 1 and cubos[i] != '.':
            continue
        else:
            return 0

    # Contar movimientos para crear la secuencia correcta
    for i in range(N):
        if i % 2 == 0 and cubos[i] != 'B':
            # Si el cubo debería tener una bola pero está vacío, contar un movimiento
            movimientos += 1
        elif i % 2 == 1 and cubos[i] != '.':
            # Si el cubo debería estar vacío pero tiene una bola, contar un movimiento
            movimientos += 1

    # Verificar si la cantidad de movimientos es válida
    if (movimientos + N // 2) % 2 == 1:
        return -1

    # Devolver la cantidad mínima de movimientos necesarios
    return movimientos // 2
```

```python
# Ejemplos de uso:
cubos1 = "..B...B.BB"
cubos2 = "BB.B.BBB..."
cubos3 = ".BBB.B"
cubos4 = "......B.B"

print(solution(cubos1))  # Debería devolver 2
print(solution(cubos2))  # Debería devolver 4
print(solution(cubos3))  # Debería devolver -1
print(solution(cubos4))  # Debería devolver 0
```
